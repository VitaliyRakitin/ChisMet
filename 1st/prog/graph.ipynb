```python
import pandas as pd
import numpy as np
import pylab as pl
%pylab inline
read_df = pd.read_csv("data_-7.csv",sep = ",")

t = read_df[[0]].values
x = read_df[[1]].values
y = read_df[[2]].values

pl.plot(t, x)
pl.axis([0, 70 , -2 , 2])
pl.xlabel("t")
pl.ylabel("x")
savefig('01t-x.pdf')

pl.show()
```

```python
pl.plot(t,y)
pl.axis([0, 70, -2 , 2])
pl.xlabel("t")
pl.ylabel("y")
savefig('01t-y.pdf')

pl.show()
```
